```python
import matplotlib.pyplot as plt
```

```python
plt.scatter(merged_data["Murder"],merged_data["Assault"])
plt.title("Murder v Assault")
plt.xlabel("Murder")
plt.ylabel("Assault")
for i, row in merged_data.iterrows():
    plt.annotate(row['State'], (row['Murder'], row['Assault']),
                 textcoords="offset points",  # how to position the text
                 xytext=(0,0),               # distance from the point
                 ha='left')                  # horizontal alignment
plt.show()
```

```python
outliers=merged_data[merged_data["State"]=="Alaska"]
outliers
```

```python
plt.scatter(merged_data["Rape"],merged_data["UrbanPop"])
plt.scatter(outliers["Rape"], outliers["UrbanPop"], color="Red")
plt.title("Rape v UrbanPop")
plt.xlabel("Rape")
plt.ylabel("UrbanPop")
for i, row in merged_data.iterrows():
    plt.annotate(row['State'], (row['Rape'], row['UrbanPop']),
                 textcoords="offset points",  # how to position the text
                 xytext=(0,0),               # distance from the point
                 ha='left')                  # horizontal alignment
plt.show()
```

```python
merged_data_dropped=merged_data.drop(columns=["Murder","Assault"])
```

```python
merged_data_dropped
```

```python
import statsmodels.api as sm
```

```python
y=merged_data_dropped["Rape"]
x=merged_data_dropped["UrbanPop"]
X=sm.add_constant(x)
model = sm.OLS(y, X).fit()
```

```python
model.summary()
```

```python
y_pred=model.predict(X)
```

```python
plt.scatter(y,y_pred)
plt.title("Rape predicted vs actuals")
plt.xlabel("Actual rape")
plt.ylabel("Predicted rape")             # horizontal alignment
plt.show()
```

```python
plt.scatter(x,y)
plt.plot(x, y_pred, color='blue', label='Regression line')     # regression line
plt.title("Rape v UrbanPop")
plt.xlabel("Rape")
plt.ylabel("UrbanPop")                # horizontal alignment
plt.show()
```
